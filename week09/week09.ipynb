Criando um subgrafo apenas com os aeroportos do brasil
(os grafos que possuem região 'nan' são aeroportos do exterior):

````python
nosBrasil = []
for no in G.nodes():
  if G.nodes[no]['region'] == "NORTE" or G.nodes[no]['region'] == "NORDESTE" or G.nodes[no]['region'] == "SUL" or G.nodes[no]['region'] == "SUDESTE" or G.nodes[no]['region'] == "CENTRO-OESTE":
    nosBrasil.append(no)

Gbrasil = G.subgraph(nosBrasil)

print("Nós da rede do brasil: ", Gbrasil.number_of_nodes())
print("Links da rede do brasil: ", Gbrasil.number_of_edges())
````

Após criar uma representação visual, vou criar a representação matricial (mixing matrix) da imagem acima considerando a REGIÃO que o aeroporto está localizado.

````python
nx.attribute_mixing_matrix(Gbrasil,'region')
````

````python
nx.attribute_assortativity_coefficient(Gbrasil,'region')
````

O coeficiente de assortatividade dessa rede é de aproximadamente ```0.372```. Por definição, uma rede á desassortativa se o coeficiente de assortatividade está entre -1 e 0 (elementos do mesmo grupo tendem a se conectar com elementos de outros grupos) e a rede é assortativa se o coeficiente é entre 0 e 1 (elementos do mesmo grupo tendem a se conectar com elementos do mesmo grupo). Como o coeficiente dessa rede está entre 0 e 1, então a rede é assortativa. Neste exemplo em específico, isso significa que aeroportos de uma região tendem a "se conectar" com aerportos da mesma região, ou seja, os voos que saem da região sul, por exemplo, tendem a ter como destino a própria região sul.

# Requisito 2
Análise bivariada entre o grau do vértice e o número médio de vizinhos.


# criando subgrafos da rede do brasil, cada subgrafo representa uma região diferente

````python
def gerarSubgrafo(G, regiao):
  gRegiao = []

  for no in G.nodes():
    if G.nodes[no]['region'] == regiao:
      gRegiao.append(no)
  
  return G.subgraph(gRegiao)

GbrasilNorte = gerarSubgrafo(Gbrasil, 'NORTE')
GbrasilNordeste = gerarSubgrafo(Gbrasil, 'NORDESTE')
GbrasilSul= gerarSubgrafo(Gbrasil, 'SUL')
GbrasilSudeste = gerarSubgrafo(Gbrasil, 'SUDESTE')
GbrasilCentroOeste = gerarSubgrafo(Gbrasil, 'CENTRO-OESTE') 
````

# Pegando os graus médios dos vizinhos de cada região

````python
degreeNorte, avgNeighborDegreeNorte = zip(*nx.average_degree_connectivity(GbrasilNorte).items())
degreeNordeste, avgNeighborDegreeNordeste = zip(*nx.average_degree_connectivity(GbrasilNordeste).items())
degreeSul, avgNeighborDegreeSul = zip(*nx.average_degree_connectivity(GbrasilSul).items())
degreeSudeste, avgNeighborDegreeSudeste = zip(*nx.average_degree_connectivity(GbrasilSudeste).items())
degreeCentroOeste, avgNeighborDegreeCentroOeste = zip(*nx.average_degree_connectivity(GbrasilCentroOeste).items())
````

# Pegando os graus médios dos vizinhos do Brasil

````python
degreeBrasil, avgNeighborDegreeBrasil = zip(*nx.average_degree_connectivity(Gbrasil).items())
````

# Transformando os graus médios em listas

````python
degreeNorte = list(degreeNorte)
avgNeighborDegreeNorte = list(avgNeighborDegreeNorte)

degreeNordeste = list(degreeNordeste)
avgNeighborDegreeNordeste = list(avgNeighborDegreeNordeste)

degreeSul = list(degreeSul)
avgNeighborDegreeSul = list(avgNeighborDegreeSul)

degreeSudeste = list(degreeSudeste)
avgNeighborDegreeSudeste = list(avgNeighborDegreeSudeste)

degreeCentroOeste = list(degreeCentroOeste)
avgNeighborDegreeCentroOeste = list(avgNeighborDegreeCentroOeste)

degreeBrasil = list(degreeBrasil)
avgNeighborDegreeBrasil = list(avgNeighborDegreeBrasil)
````

# Análise Bivariada da rede do Brasil

````python
plt.style.use('fivethirtyeight')
fig, ax = plt.subplots(1,1,figsize=(15,15))
sb.regplot(degreeBrasil, avgNeighborDegreeBrasil,ax=ax)

ax.set_title('Análise Bivariada da rede do Brasil')
ax.set_ylabel('Grau médio dos vizinhos')
ax.set_xlabel('Grau dos nós')
ax.text(35, 100, 'Assortatividade: '+str(nx.degree_assortativity_coefficient(Gbrasil)),
        bbox={'facecolor': '#FED141', 'alpha': 0.3})
ax.set_xlim(0,65)


plt.savefig("r2_grauBrasil.png", 
            format="png",
            dpi=400, 
            bbox_inches="tight",
            transparent=True)

plt.show()
````

Em uma rede desassortativa com relação ao grau, nós de alto grau tendem a se conectar com nós de baixo grau. Em uma rede assortativa com relação ao grau, nós de alto grau tendem a se conectar com nós de alto grau. Quando observamos o gráfico ```Análise Bivariada da rede do Brasil```, observa-se que nós de grau alto tendem a se conectar com nós de grau baixo, ou seja, a rede do brasil é uma rede desassortativa em relação ao grau. Isso é confirmado ao observar os coeficientes de assortatitivade das 5 regiões do Brasil: ambos seguem a tendência da rede do brasil, pois todos são graus desassortativos. As regiões ```Sudeste``` e ```Sul```, são, respectivamente as regiões com maior grau de desassortividade. Na prática, isso signigica que essas são as regiões que mais se conectam com outras regiões, em particular, com as que menos se conectam com outras regiões (```Norte``` e ```Nordeste```).
